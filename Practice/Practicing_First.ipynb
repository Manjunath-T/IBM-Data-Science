```python
set_1 = {'A', 'B', 'C'}
set_2 = {'D', 'A', 'B'}

#INtersection [Common Elements]
set_i = set_1 & set_2 
print(set_i )

#Union [All Elements]
set_u = set_1.union(set_2)
print( set_u)

#difference
print(set_1.difference(set_2))
print(set_2.difference(set_1))

#subset
print(set_1.issubset(set_u))
print(set_2.issubset(set_u))

#superset
print(set_u.issuperset(set_1))
print(set_u.issuperset(set_2))

tuple_u = (1,2,3,4,5,6,)

print(tuple_u[:9])
```

```python
class Points(object):
  def __init__(self,x,y):

    self.x=x
    self.y=y

  def print_point(self):

    print('x=',self.x,' y=',self.y)

p2=Points(1,2)

p2.x=3

p2.print_point()
```
